## Глава 1. Быстрый старт

```python
# независимая х и зависимая у переменные
x = np.linspace(0, 10, 50)
y = x
```

```python
# построение графика
plt.title('Линейная зависимость y = x') # заголовок
plt.xlabel('x') # ось абцисс
plt.ylabel('y') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y)
```

```python
# изменим цвет линии и ее тип
# построение графика
plt.title('Линейная зависимость y = x') # заголовок
plt.xlabel('x') # ось абцисс
plt.ylabel('y') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y, 'r--')
```

```python
# построим несколько графиков на одном поле
# линейная  зависимость
x = np.linspace(0, 10, 50)
y1 = x

# квадратичная зависимость
y2 = [i**2 for i in x]

# построение графика
plt.title('Зависимости: y1 = x, y2 = x^2') # заголовок
plt.xlabel('x')
plt.ylabel('y1, y2')
plt.grid()
plt.plot(x, y1, x, y2) # построение графика
```

```python
# построим несколько графиков на нескольких полях
# линейная  зависимость
x = np.linspace(0, 10, 50)
y1 = x

# квадратичная зависимость
y2 = [i**2 for i in x]

# построение графиков
plt.figure(figsize=(9, 9)) # общее поле графиков

plt.subplot(2, 1, 1) # местоположение графика 1
plt.plot(x, y1)
plt.title('Зависимости: у1 =х, у2 = х^2') # заголовок
plt.ylabel('y1', fontsize=14) # ось ординат
plt.grid(True)
plt.subplot(2, 1, 2) # местоположение графика 2
plt.plot(x, y2) 
plt.xlabel('x', fontsize=14)
plt.ylabel('y2', fontsize=14)
plt.grid()
```

```python
# построение диаграммы
fruits = ['apple', 'peach', 'orange', 'bannana', 'melon']
counts = [34, 25, 43, 31, 17]
plt.bar(fruits, counts) # построение диаграммы
plt.title('Fruits!')
plt.xlabel('Fruit')
plt.ylabel('Count')
```

```python
from matplotlib.ticker import MultipleLocator, \
    FormatStrFormatter, AutoMinorLocator

x = np.linspace(0, 10, 10)
y1 = 4*x
y2 = [i**2 for i in x]

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Графики зависимостей: y1=4*x, y2=x^2', 
            fontsize=16)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y1, y2', fontsize=14)
ax.grid(which='major', linewidth=1.2)
ax.grid(which='minor', linestyle='--', color='gray', 
       linewidth=0.5)
ax.scatter(x, y1, c='red', label='y1 = 4*x')
ax.plot(x, y2, label='y2 = x^2')
ax.legend()

ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='major', length=10, width=2)
ax.tick_params(which='minor', length=5, width=1)

plt.show()
```

## Глава 3. Настройка графиков

### Урок 3.1 Работа с легендой

```python
# первый вариант
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]
plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend()
```

```python
# второй вариант
plt.plot(x, y1, 'o-r')
plt.plot(x, y2, 'o-.g')
plt.legend(['L1', 'L2'])
```

```python
# третий вариант
line1, = plt.plot(x, y1, 'o-b')
line2, = plt.plot(x, y2, 'o-.m')
plt.legend((line2, line1), ['L2', 'L1'])
```

```python
locs = ['best', 'upper right', 'upper left', 'lower left', 
        'lower right', 'right', 'center left', 
        'center right', 'lower center', 'upper center',
        'center']
plt.figure(figsize=(12, 12))
for i in range(3):
    for j in range(4):
        if i*4+j < 11:
            plt.subplot(3, 4, i*4+j+1)
            plt.title(locs[i*4+j])
            plt.plot(x, y1, 'o-r', label='line 1')
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i*4+j])
        else:
            break
```

```python
# более гибкая настройка legend
# bbox_to_anchor = (x, y, width, height)
# bbox_to anchor(x, y)
plt.plot(x, y1, 'o-r', label='line1')
plt.plot(x, y2, 'o-.g', label='line2')
plt.legend(bbox_to_anchor=(1, 0.6))
```

```python
# пример работы с параметрами
plt.plot(x, y1, 'o-r', label='line1')
plt.plot(x, y2, 'o-.g', label='line2')
plt.legend(fontsize=12, shadow=True, framealpha=0.5, 
           facecolor='xkcd:aqua', edgecolor='black',
           title='Легенда')
```

### Урок 3.2 Компановка графиков

```python
# рассмотрим инструмент GridSpec
import matplotlib.gridspec as gridspec

x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]

fg = plt.figure(figsize=(7, 3), constrained_layout=True)
gs = gridspec.GridSpec(ncols=2, nrows=1, figure=fg)
# альтернативный вариант создания gridspec
# gs = fg.add_gridspec(1, 2)

fig_ax_1 = fg.add_subplot(gs[0, 0])
plt.plot(x, y1)

fig_ax_2 = fg.add_subplot(gs[0, 1])
plt.plot(x, y2)
```

```python
# построим графики в новой компановке
x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]
y3 = [-7, -4, 2, -4, -7]

fg = plt.figure(figsize=(9, 4), constrained_layout=True)
gs = fg.add_gridspec(2, 2)

fig_ax_1 = fg.add_subplot(gs[0, :])
plt.plot(x, y2)

fig_ax_2 = fg.add_subplot(gs[1, 0])
plt.plot(x, y3)

fig_ax_3 = fg.add_subplot(gs[1, 1])
plt.plot(x, y3)
```

```python
# еще несколько возможностей в компановке
fg = plt.figure(figsize=(15, 20), constrained_layout=True)
gs = fg.add_gridspec(5, 5)

fig_ax_1 = fg.add_subplot(gs[0, :3])
fig_ax_1.set_title('gs[0, :3]')

fig_ax_2 = fg.add_subplot(gs[0, 3:])
fig_ax_2.set_title('gs[0, 3:]')

fig_ax_3 = fg.add_subplot(gs[1:, 0])
fig_ax_3.set_title('gs[1:, 0]')

fig_ax_4 = fg.add_subplot(gs[1:, 1])
fig_ax_4.set_title('gs[1:, 1]')

fig_ax_5 = fg.add_subplot(gs[1, 2:])
fig_ax_5.set_title('gs[1, 2:]')
fig_ax_6 = fg.add_subplot(gs[2:4, 2])
fig_ax_6.set_title('gs[2:4, 2]')
fig_ax_7 = fg.add_subplot(gs[2:4, 3:])
fig_ax_7.set_title('gs[2:4, 3:]')
fig_ax_8 = fg.add_subplot(gs[4, 3:])
fig_ax_8.set_title('gs[4, 3:]')
```

```python
# можно задаться размерами областей и после передать
fg = plt.figure(figsize=(5, 5),constrained_layout=True)
widths = [1, 3]
heights = [2, 0.7]

gs = fg.add_gridspec(ncols=2, nrows=2, width_ratios=widths,
                     height_ratios=heights)

fig_ax_1 = fg.add_subplot(gs[0, 0])
fig_ax_1.set_title('w:1, h:2')

fig_ax_2 = fg.add_subplot(gs[0, 1])
fig_ax_2.set_title('w:3, h:2')

fig_ax_3 = fg.add_subplot(gs[1, 0])
fig_ax_3.set_title('w:1, h:0.7')

fig_ax_4 = fg.add_subplot(gs[1, 1])
fig_ax_4.set_title('w:3, h:0.7')
```

### Урок 3.3 Текстовые элементы графика

```python
# аннотация - простой пример
import math
x = list(range(-5, 6))
y = [i**2 for i in x]

plt.annotate('min', xy=(0, 0), xycoords='data',
            xytext=(0, 10), textcoords='data',
            arrowprops=dict(facecolor='g'))
plt.plot(x, y)
```

```python
# возможности работы с параметром connectionstyle
fig, axs = plt.subplots(2, 3, figsize=(12, 7))

conn_style = [ "angle,angleA=90,angleB=0,rad=0.0",
    "angle3,angleA=90,angleB=0",
    "arc,angleA=0,angleB=0,armA=0,armB=40,rad=0.0",
    "arc3,rad=-1.0",
    "bar,armA=0.0,armB=0.0,fraction=0.1,angle=70",
    "bar,fraction=-0.5,angle=180"]

for i in range(2):
    for j in range(3):
        axs[i, j].text(0.1, 0.5, 
            '\n'.join(conn_style[i*3+j].split(',')))
        axs[i, j].annotate('text', xy=(0.2, 0.2), 
            xycoords='data', xytext=(0.7, 0.8), 
            textcoords='data',
            arrowprops=dict(arrowstyle='->',
                connectionstyle=conn_style[i*3+j]))
```

### Урок 3.5 Цветовая полоса Colorbar

```python
# построим colorbar
np.random.seed(123)
vals = np.random.randint(10, size=(7, 7))

plt.pcolor(vals)
plt.colorbar()
```

```python
# дискретное разделение цветов на полоске
np.random.seed(123)
vals = np.random.randint(10, size=(7, 7))

plt.pcolor(vals, cmap=plt.get_cmap('viridis', 11))
plt.colorbar()
```

```python
# более тонкая настройка с помощью insert_axes
import mpl_toolkits
from mpl_toolkits.axes_grid1 import inset_locator as ins

np.random.seed(123)
vals = np.random.randint(10, size=(7, 7))

fig, ax = plt.subplots()

gr = ax.pcolor(vals)

axins = ins.inset_axes(ax, width='7%', height='50%', 
                   loc='lower left', 
                   bbox_to_anchor=(1.05, 0., 1, 1),
                   bbox_transform=ax.transAxes,
                   borderpad=0)
plt.colorbar(gr, cax=axins)
```

```python
# модифицируем шкалу и устанавливаем надпись
import mpl_toolkits
from mpl_toolkits.axes_grid1 import inset_locator as ins

np.random.seed(123)
vals = np.random.randint(10, size=(7, 7))

fig, ax = plt.subplots()

gr = ax.pcolor(vals)

axins = ins.inset_axes(ax, width='7%', height='50%', 
                   loc='lower left', 
                   bbox_to_anchor=(1.05, 0., 1, 1),
                   bbox_transform=ax.transAxes,
                   borderpad=0)
plt.colorbar(gr, cax=axins, ticks=[0, 5, 10],
             label='Value')
```

```python
# установим текстовые надписи на шкале
np.random.seed(123)
vals = np.random.randint(10, size=(7, 7))

fig, ax = plt.subplots()

gr = ax.pcolor(vals)

axins = ins.inset_axes(ax, width='7%', height='50%', 
                   loc='lower left', 
                   bbox_to_anchor=(1.05, 0., 1, 1),
                   bbox_transform=ax.transAxes,
                   borderpad=0)
cbar = plt.colorbar(gr, cax=axins, ticks=[0, 5, 9],
                   label='Value')
cbar.ax.set_yticklabels(['Low', 'Medium', 'High'])
```

```python
# дополнительные настройки colorbar
np.random.seed(123)
vals = np.random.randint(10, size=(7, 7))

plt.pcolor(vals, cmap=plt.get_cmap('viridis', 11))
plt.colorbar(orientation='horizontal', shrink=0.9, 
            extend='max', extendfrac=0.2,
            extendrect=False, drawedges=False)
```
