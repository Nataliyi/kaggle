## Глава 1. Быстрый старт

```python
# независимая х и зависимая у переменные
x = np.linspace(0, 10, 50)
y = x
```

```python
# построение графика
plt.title('Линейная зависимость y = x') # заголовок
plt.xlabel('x') # ось абцисс
plt.ylabel('y') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y)
```

```python
# изменим цвет линии и ее тип
# построение графика
plt.title('Линейная зависимость y = x') # заголовок
plt.xlabel('x') # ось абцисс
plt.ylabel('y') # ось ординат
plt.grid() # включение отображения сетки
plt.plot(x, y, 'r--')
```

```python
# построим несколько графиков на одном поле
# линейная  зависимость
x = np.linspace(0, 10, 50)
y1 = x

# квадратичная зависимость
y2 = [i**2 for i in x]

# построение графика
plt.title('Зависимости: y1 = x, y2 = x^2') # заголовок
plt.xlabel('x')
plt.ylabel('y1, y2')
plt.grid()
plt.plot(x, y1, x, y2) # построение графика
```

```python
# построим несколько графиков на нескольких полях
# линейная  зависимость
x = np.linspace(0, 10, 50)
y1 = x

# квадратичная зависимость
y2 = [i**2 for i in x]

# построение графиков
plt.figure(figsize=(9, 9)) # общее поле графиков

plt.subplot(2, 1, 1) # местоположение графика 1
plt.plot(x, y1)
plt.title('Зависимости: у1 =х, у2 = х^2') # заголовок
plt.ylabel('y1', fontsize=14) # ось ординат
plt.grid(True)
plt.subplot(2, 1, 2) # местоположение графика 2
plt.plot(x, y2) 
plt.xlabel('x', fontsize=14)
plt.ylabel('y2', fontsize=14)
plt.grid()
```

```python
# построение диаграммы
fruits = ['apple', 'peach', 'orange', 'bannana', 'melon']
counts = [34, 25, 43, 31, 17]
plt.bar(fruits, counts) # построение диаграммы
plt.title('Fruits!')
plt.xlabel('Fruit')
plt.ylabel('Count')
```

```python
from matplotlib.ticker import MultipleLocator, \
    FormatStrFormatter, AutoMinorLocator

x = np.linspace(0, 10, 10)
y1 = 4*x
y2 = [i**2 for i in x]

fig, ax = plt.subplots(figsize=(8, 6))
ax.set_title('Графики зависимостей: y1=4*x, y2=x^2', 
            fontsize=16)
ax.set_xlabel('x', fontsize=14)
ax.set_ylabel('y1, y2', fontsize=14)
ax.grid(which='major', linewidth=1.2)
ax.grid(which='minor', linestyle='--', color='gray', 
       linewidth=0.5)
ax.scatter(x, y1, c='red', label='y1 = 4*x')
ax.plot(x, y2, label='y2 = x^2')
ax.legend()

ax.xaxis.set_minor_locator(AutoMinorLocator())
ax.yaxis.set_minor_locator(AutoMinorLocator())
ax.tick_params(which='major', length=10, width=2)
ax.tick_params(which='minor', length=5, width=1)

plt.show()
```

```python
# второй вариант
plt.plot(x, y1, 'o-r')
plt.plot(x, y2, 'o-.g')
plt.legend(['L1', 'L2'])
```

```python
# третий вариант
line1, = plt.plot(x, y1, 'o-b')
line2, = plt.plot(x, y2, 'o-.m')
plt.legend((line2, line1), ['L2', 'L1'])
```

```python
locs = ['best', 'upper right', 'upper left', 'lower left', 
        'lower right', 'right', 'center left', 
        'center right', 'lower center', 'upper center',
        'center']
plt.figure(figsize=(12, 12))
for i in range(3):
    for j in range(4):
        if i*4+j < 11:
            plt.subplot(3, 4, i*4+j+1)
            plt.title(locs[i*4+j])
            plt.plot(x, y1, 'o-r', label='line 1')
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i*4+j])
        else:
            break
```

```python
# пример работы с параметрами
plt.plot(x, y1, 'o-r', label='line1')
plt.plot(x, y2, 'o-.g', label='line2')
plt.legend(fontsize=12, shadow=True, framealpha=0.5, 
           facecolor='xkcd:aqua', edgecolor='black',
           title='Легенда')
```

### Урок 3.2 Компановка графиков

```python
# рассмотрим инструмент GridSpec
import matplotlib.gridspec as gridspec

x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]

fg = plt.figure(figsize=(7, 3), constrained_layout=True)
gs = gridspec.GridSpec(ncols=2, nrows=1, figure=fg)
# альтернативный вариант создания gridspec
# gs = fg.add_gridspec(1, 2)

fig_ax_1 = fg.add_subplot(gs[0, 0])
plt.plot(x, y1)

fig_ax_2 = fg.add_subplot(gs[0, 1])
plt.plot(x, y2)
```

```python
# построим графики в новой компановке
x = [1, 2, 3, 4, 5]
y1 = [9, 4, 2, 4, 9]
y2 = [1, 7, 6, 3, 5]
y3 = [-7, -4, 2, -4, -7]

fg = plt.figure(figsize=(9, 4), constrained_layout=True)
gs = fg.add_gridspec(2, 2)

fig_ax_1 = fg.add_subplot(gs[0, :])
plt.plot(x, y2)

fig_ax_2 = fg.add_subplot(gs[1, 0])
plt.plot(x, y3)

fig_ax_3 = fg.add_subplot(gs[1, 1])
plt.plot(x, y3)
```

```python
# еще несколько возможностей в компановке
fg = plt.figure(figsize=(15, 20), constrained_layout=True)
gs = fg.add_gridspec(5, 5)

fig_ax_1 = fg.add_subplot(gs[0, :3])
fig_ax_1.set_title('gs[0, :3]')

fig_ax_2 = fg.add_subplot(gs[0, 3:])
fig_ax_2.set_title('gs[0, 3:]')

fig_ax_3 = fg.add_subplot(gs[1:, 0])
fig_ax_3.set_title('gs[1:, 0]')

fig_ax_4 = fg.add_subplot(gs[1:, 1])
fig_ax_4.set_title('gs[1:, 1]')

fig_ax_5 = fg.add_subplot(gs[1, 2:])
fig_ax_5.set_title('gs[1, 2:]')
fig_ax_6 = fg.add_subplot(gs[2:4, 2])
fig_ax_6.set_title('gs[2:4, 2]')
fig_ax_7 = fg.add_subplot(gs[2:4, 3:])
fig_ax_7.set_title('gs[2:4, 3:]')
fig_ax_8 = fg.add_subplot(gs[4, 3:])
fig_ax_8.set_title('gs[4, 3:]')
```

```python
# можно задаться размерами областей и после передать
fg = plt.figure(figsize=(5, 5),constrained_layout=True)
widths = [1, 3]
heights = [2, 0.7]

gs = fg.add_gridspec(ncols=2, nrows=2, width_ratios=widths,
                     height_ratios=heights)

fig_ax_1 = fg.add_subplot(gs[0, 0])
fig_ax_1.set_title('w:1, h:2')

fig_ax_2 = fg.add_subplot(gs[0, 1])
fig_ax_2.set_title('w:3, h:2')

fig_ax_3 = fg.add_subplot(gs[1, 0])
fig_ax_3.set_title('w:1, h:0.7')

fig_ax_4 = fg.add_subplot(gs[1, 1])
fig_ax_4.set_title('w:3, h:0.7')
```

```python
# возможности работы с параметром connectionstyle
fig, axs = plt.subplots(2, 3, figsize=(12, 7))

conn_style = [ "angle,angleA=90,angleB=0,rad=0.0",
    "angle3,angleA=90,angleB=0",
    "arc,angleA=0,angleB=0,armA=0,armB=40,rad=0.0",
    "arc3,rad=-1.0",
    "bar,armA=0.0,armB=0.0,fraction=0.1,angle=70",
    "bar,fraction=-0.5,angle=180"]

for i in range(2):
    for j in range(3):
        axs[i, j].text(0.1, 0.5, 
            '\n'.join(conn_style[i*3+j].split(',')))
        axs[i, j].annotate('text', xy=(0.2, 0.2), 
            xycoords='data', xytext=(0.7, 0.8), 
            textcoords='data',
            arrowprops=dict(arrowstyle='->',
                connectionstyle=conn_style[i*3+j]))
```

```python
# изменим правила заливки
plt. plot(x, y, c='r')
plt.fill_between(x, y, where=(y > 0.75) | (y < -0.75))
```

```python
# заливка области между 0 и y
plt.plot(x, y, c='r')
plt.fill_between(x, y, where = (y > 0))
```

```python
# заливка областей между 0.5 и x
plt.plot(x, y, c='r')
plt.grid()
plt.fill_between(x, 0.5,  y, where = (y >= 0.5))
```

```python
# заливка областей между y и 1
plt.plot(x, y, c='r')
plt.grid()
plt.fill_between(x, y, 1)
```

```python
# там, где точек выше 500, их уже не видно
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

# зададим интервал маркеров markevery. График до и после
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
ax[0].plot(x, y, marker='o', c='g')
ax[1].set_title('Каждая 30 точка', fontsize=14)
ax[1].plot(x, y, marker='o', c='g', markevery=30, 
                  ms=11)
```

```python
# возможности раставления маркеров с помощью markevery
x = np.arange(0.0, 5, 0.01)
y = np.cos(x * np.pi)

m_ev_case = [None, 
             10, 
             (100, 30), 
             slice(100, 400, 15),
             [0, 100, 200, 300],
            [10, 50, 100]]
fig, ax = plt.subplots(2, 3, figsize=(10, 7))
axs = [ax[i, j] for i in range(2) for j in range(3)]
for i, case in enumerate(m_ev_case):
    axs[i].set_title(case)
    axs[i].plot(x, y, 'o', ls='-', ms=7, markevery=case)
```

```python
# обрезка графика c помощью numpy.ma.masked_where
x = np.arange(0.0, 5, 0.01)
y = np.cos(x * np.pi)

y_masked = np.ma.masked_where(y < -0.5, y)

plt.ylim(-1, 1)
plt.plot(x, y_masked, linewidth=3)
```

### Урок 4.2 Ступенчатый, стековый, точечный и другие графики

```python
# ступенчатый график
x = np.arange(0, 7)
y = x

where_set = ['pre', 'post', 'mid']
fig, axs = plt.subplots(1, 3, figsize=(15, 4))
for i, ax in enumerate(axs):
    # строим ступенчатый график с помощью функции step
    ax.step(x, y, 'g-o', where=where_set[i])
    ax.grid()
```

```python
# стековый график
x = np.arange(0, 11, 1)
y1 = np.array([(-0.2)*i**2+2*i for i in x])
y2 = np.array([(-0.4)*i**2+4*i for i in x])
y3 = np.array([2*bi for i in x])

labels = ['y1', 'y2', 'y3']

fig, ax = plt.subplots()
# строим стековый график с помощью функции stackplot
ax.stackplot(x, y1, y2, y3, labels=labels)
ax.legend(loc='upper left')
```

```python
# stem-график
x = np.arange(0, 10.5, 0.5)
y = np.array([(-0.2)*i**2+2*i for i in x])

# строим stem-график с помощью функции stem
plt.stem(x, y)
```

```python
# реализуем пример stem графика с доп параметрами
plt.stem(x, y, linefmt='b.', markerfmt='p', bottom = 0.5,
        basefmt='r-.')
# bottom - это y координата базовой линии
```

```python
# точечный график
x = np.arange(0, 10.5, 0.5)
y = np.cos(x)

# строим точечный график с помощью функции scatter
plt.scatter(x, y)
```

```python
# исп расширенные параметры для настройки точечного графика
plt.scatter(x, y, s=80, c='r', marker='D', 
           lw=2, edgecolors='g')
```

```python
# пример, демонстрирующий работу с цветом
import matplotlib.colors as mcolors

bc = mcolors.BASE_COLORS
x = np.arange(0, 10.5, 0.25)
y = np.cos(x)

num_set = np.random.randint(1, len(bc), len(x))
sizes = num_set * 35
colors = [list(bc.keys())[i] for i in num_set]
plt.scatter(x, y, s=sizes, alpha=.4, c=colors, lw=2,
           edgecolors='face')
plt.plot(x, y, 'g--', alpha=.4)
```

### Урок 4.3 Столбчатые и круговые диаграммы

```python
# построим простую диаграмму
np.random.seed(123)

groups = [f'P{i}' for i in range(7)]
counts = np.random.randint(3, 10, len(groups))
plt.bar(groups, counts)
```

```python
# если заменим bar на barh, то получим горизонтальную 
plt.barh(groups, counts)
```

```python
# построим более сложный пример с доп параметрами
import matplotlib.colors as mcolors

bc = mcolors.BASE_COLORS
np.random.seed(123)

groups = [f'P{i}' for i in range(7)]
counts = np.random.randint(3, 10, len(groups))
# зададим в списке ширину столбцов
width = counts * 0.1
colors = [['r', 'b', 'g'][int(
    np.random.randint(0, 3, 1))] for _ in counts]
plt.bar(groups, counts, width=width, alpha=.6, bottom=2,
       color=colors, edgecolor='k', lw=2)
```

```python
# групповые столбчатые диаграммы
cat_par = [f'P{i}' for i in range(5)]

g1 = [10, 34, 85, 16, 77]
g2 = [9, 4, 18, 10, 65]

x = np.arange(len(cat_par))

width = 0.3

fig, ax = plt.subplots()

rects1 = ax.bar(x - width/2, g1, width=width, label='g1')
rects2 = ax.bar(x + width/2, g2, width=0.3, label='g2')

ax.set_title('Пример групповой диаграммы')
ax.set_xticks(x)
ax.set_xticklabels(cat_par)
ax.legend()
```

```python
# диаграмма с errrorbar элементом
np.random.seed(123)
rnd = np.random.randint

cat_par = [f'P{i}' for i in range(5)]
g1 = [10, 21, 34, 17, 12]
error = np.array(
    [[rnd(2, 7), rnd(2, 7)] for _ in range(len(cat_par))]).T

fig, axs = plt.subplots(1, 2, figsize=(10 ,5))
axs[0].bar(cat_par, g1, yerr=5, ecolor='r',
          alpha=.5, edgecolor='b', lw=2)
axs[1].bar(cat_par, g1, yerr=error, ecolor='k',
          alpha=.5, edgecolor='k', lw=2)
```

```python
# классическая круговая диаграммма
vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'Ferrary', 'BMW', 'Renault']

fig, ax = plt.subplots()
# строим круговую диаграмму с помощью pie
ax.pie(vals, labels=labels)
ax.axis('equal')
```

```python
# cоздадим пример с настройкой некоторых параметров pie
vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'Ferrary', 'BMW', 'Renault']

explode = (0.1, 0, 0.15, 0, 0)

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, autopct='%1.1f%%', shadow=True,
      explode=explode, wedgeprops={'lw': 2, 
        'ls': '--', 'edgecolor': 'k'}, 
      rotatelabels=True)
# установим автоматическое масштабирование графика
ax.axis('auto')
```

```python
# вложенные круговые диаграммы
fig, ax = plt.subplots()

offset = 0.4

data = np.array([[5, 10, 7], [8, 15, 5], [11, 9, 7]])
cmap = plt.get_cmap('tab20b')
b_colors = cmap(np.array([0 ,8, 12]))
sm_colors = cmap(np.array([1, 2, 3, 9, 10, 11, 13, 14, 15]))

ax.pie(data.sum(axis=1), radius=1, colors=b_colors,
      wedgeprops=dict(width=offset, edgecolor='w'))
ax.pie(data.flatten(), radius=1-offset, 
      colors=sm_colors, wedgeprops=dict(width=offset,
                                       edgecolor='w'));
# точка запятая в конце строки убирает весь лишний текст
```

```python
# круговая диаграмма в виде бублика
vals = [24, 17, 53, 21, 35]
labels = ['Putin', 'Timoshenko', 'Merkel', 'Al-Babi', 
          'Groooot']

fig, ax = plt.subplots()
ax.pie(vals, labels=labels, wedgeprops=dict(width=.5));
```

```python
# построим пример с расширенными параметрами
fig, axs = plt.subplots(1, 2, figsize=(10, 3))

p1 = axs[0].imshow(data, cmap='winter', aspect='equal',
        vmin=-1, vmax=1, origin='lower')
fig.colorbar(p1, ax=axs[0], aspect=5)

p2 = axs[1].imshow(data, cmap='plasma', aspect='equal',
        origin='lower', extent=(0, 30, 0, 30))
fig.colorbar(p2, ax=axs[1], aspect=10);
```

## Глава 5. Построение 3D графиков

```python
# линейный 3D график plot
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-np.pi, np.pi, 50)
y = x
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, label='parametric curve');
```

```python
# точечный 3D график scatter
x = np.random.randint(-5, 5, 40)
y = np.random.randint(0, 10, 40)
z = np.random.randint(-5, 5, 40)
s = np.random.randint(10, 100, 40)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, s=s);
```

```python
# каркасная поверхность plot_wireframe()
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u) * np.sin(v)
y = np.sin(u) * np.sin(v)
z = np.cos(v)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_wireframe(x, y, z);
```

```python
# поверхность plot_surface
u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='inferno');
```
