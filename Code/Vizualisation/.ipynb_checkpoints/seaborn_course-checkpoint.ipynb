## Глава 1. Эстетическая фигура

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# построим график с настройками matplotlib
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i * .5)*(7 - i) * flip)
        
sinplot()
```

```python
# для настроек по умолчанию seaborn используем set()
import seaborn as sb

sb.set()
sinplot()
```

```python
# поменяем тему сюжета seaborn
sb.set_style('whitegrid')
sinplot()
```

```python
# удалим верхние оси и сменим сетку
sb.set_style('white')
sinplot()
sb.despine()
```

```python
# посмотрим на доступные параметры set_style()
sb.axes_style()
```

```python
# наложим сетку сверху графика
sb.set_style('darkgrid', {'axes.axisbelow': False})
sb.despine()
sinplot()
```

```python
# посмотрим на др масштаб графика с помощью set_context()
sb.set_context('talk')
sb.set_style()
sinplot()
```

## Глава 2. Цветовая палитра

```python
# категориальные цветовые палитры
current_palette = sb.color_palette()
# palplot исп для горизонтального построения массива цветов
sb.palplot(current_palette)
```

```python
# последовательные цветовые палитры
current_palette = sb.color_palette('Greens')
sb.palplot(current_palette)
```

```python
# расходящаяся палитра
current_palette = sb.color_palette('BrBG', 7)
sb.palplot(current_palette)
```

```python
# мы можем изменять палитру с помощью set_palette
sb.set_style('white')
sb.set_palette('husl')
sb.set_context('paper')
sinplot()
sb.despine()
```

## Глава 3. Гистограммы

### Урок 3.1 Displot

```python
# используем pandas для получения данных
import pandas as pd

df = sb.load_dataset('iris')
sb.displot(df['petal_length'])
```

```python
# оценка плотности и непрерывности данных
sb.displot(df['petal_length'], kind='kde')
```

### Урок 3.2 Jointplot

```python
# oбъединим два графика с помощью joinplot
df = sb.load_dataset('iris')
sb.jointplot(x='petal_length', y='petal_width', data=df)
```

```python
# гексагональное объединение
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length', y = 'petal_width', 
           data=df, kind='hex')
```

```python
# оценка плотности ядра
df = sb.load_dataset('iris')
sb.jointplot(x = 'petal_length', y = 'petal_width', 
           data=df, kind='kde')
```

## Глава 4. Визуализация парных отношений

### Урок 4.1 Pairplot

```python
# построим график парных отношений pairplot
df = sb.load_dataset('iris')
sb.set_style('ticks')
sb.pairplot(df, hue='species', diag_kind='kde',
           kind='scatter', palette='husl')
```

### Урок 4.2 PairGrid

```python
# другой способ построения парных отношений - pairgrid
df = sb.load_dataset('iris')
g = sb.PairGrid(df)
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

```python
# настроим цвет графиков исп все категориальные переменные
df = sb.load_dataset('iris')
g = sb.PairGrid(df, hue='species')
g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)
```

## Глава 7. Мультипанельные категориальные графики

### Урок 7.1 Catplot

```python
# построим график на сетке FacetGrid
df = sb.load_dataset('exercise')
sb.catplot(x='time', y='pulse', hue='kind', data=df,
          kind='point')
```

```python
# изменим тип графика и представим на тех же данных
df = sb.load_dataset('exercise')
sb.catplot(x='time', y='pulse', hue='kind', data=df,
          kind='violin')
```
